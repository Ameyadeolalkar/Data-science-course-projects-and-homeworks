```python
import pandas as pd
import json
import matplotlib as plt
%matplotlib inline
```

```python
with open('train.json','rU') as f:
    data = json.load(f)
```

```python
cuisines = pd.DataFrame(data)
```

```python
cuisines.head()
```

```python
ingredient = pd.DataFrame(i for i in cuisines['ingredients'])
```

```python
ingredient.head()
```

```python
ingredient['id']=cuisines['id']
```

```python
ingredient.head()
```

```python
cuisine_ingredients = pd.merge(cuisines,ingredient)
```

```python
cuisine_ingredients.head()
```

```python
length = []
for i in cuisine_ingredients['ingredients']:
    length.append(len(i))
```
